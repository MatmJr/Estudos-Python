```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv" #titanic data set
data = pd.read_csv(url)
```

```python
data.head() # 5 primeiras linhas
```

```python
data = data.drop(["Name"], axis = 1) #retirar variável 
data.head()
```

```python
#meu data set tem uma coluna a menos, então preciso criar a coluna PassengerId

n =  np.arange(1,888)
data['PassengerId'] = n
data
```

```python
#tornar a coluna PassengerId índice e renomear a coluna Survived
data = data.set_index('PassengerId')
data = data.rename(columns = {'Survived': 'target'}, inplace = False)
data
```

Descritiva

```python
data.describe()
```

```python
data.describe(include=['O'])
```

```python
data['Sex_Flag'] = np.where(data['Sex'] == 'female',1, 0)

data['Pclass_1'] = np.where(data['Pclass'] == 1, 1, 0)
data['Pclass_2'] = np.where(data['Pclass'] == 2, 1, 0)
data['Pclass_3'] = np.where(data['Pclass'] == 3, 1, 0)
data
```

```python
data = data.drop(['Sex','Pclass'], axis=1)
```

```python
data.head()
```

```python
#procurar valores nulos no dataset
data.isnull().sum()
# se houver missings usar
data.fillna(0,inplace=True)
```

Amostragem

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop(['target'],axis=1),
                                                     data['target'],
                                                     test_size = 0.3,
                                                     random_state = 1234)

[{'treino':x_train.shape},{'teste':x_test.shape}]
```

Modelo - Random Forest

```python
rf = RandomForestClassifier(n_estimators=1000,
                            criterion='gini',
                            max_depth = 5)

rf.fit(x_train,y_train)
```

```python
prob = rf.predict_proba(data.drop('target',axis =1))[:,1]
cla = rf.predict(data.drop('target',axis = 1))
```

```python
data['Probability'] = prob
data['Classification'] = cla
```
